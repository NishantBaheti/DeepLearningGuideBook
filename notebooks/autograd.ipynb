## Differentiability

* A function is really differentiable when it has a well defined derivative at a given point
* In simpler terms if function has a slope or rate of exchange than can be calculated (must not have any sharp edges)
* Differentiable $f(x) = x^2$
* Non differentiable $f(x) = |x|$

\begin{align*}
f'(x) &= \lim_{h \rightarrow 0} \frac{f(x + h) - f(x)}{h} \\\\
&\text{or} \\
f'(x) &= \lim_{h \rightarrow 0} \frac{f(x + h) - f(x - h)}{2h} \\\\
\end{align*}

```python
def differentiable(x):
    return x**2

def non_differentiable(x):
    return np.abs(x)

def derivative(func, x):
    h = 1e-5
    return (func(x + h) - func(x - h)) / (2.0 * h)
```

```python
x = 1
h = 1e-10
f_x = differentiable(x)
f_xh = differentiable(x + h)

df_x = (f_xh - f_x) / h
print(df_x)

print(derivative(differentiable, x))
```

```python
def plot_differentiation(func):
    x = np.arange(-10, 10.1, 0.1)
    h = 1e-5
    f_x_plus_h = func(x + h)
    f_x_minus_h = func(x - h)

    df_x = (f_x_plus_h - f_x_minus_h) / (2 * h)

    fig, ax = plt.subplots(1, 2, figsize=(5, 2))
    ax[0].plot(x, f_x_plus_h, label='f(x)')
    ax[0].plot(x, f_x_minus_h, label='f(x+h)')
    # Plot the derivative
    ax[1].plot(x, df_x, label='df(x)/dx')
    fig.tight_layout()
    fig.show();
```

```python
plot_differentiation(differentiable)
```

```python
plot_differentiation(non_differentiable)
```
