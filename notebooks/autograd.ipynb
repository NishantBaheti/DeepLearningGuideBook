## Differenciability

* A function is really differenciable when it has a well defined derivative at a given point
* In simpler terms if function has a slope or rate of exchange than can be calculated (must not have any sharp edges)
* Differenciable $f(x) = x^2$
* Non differenciable $f(x) = |x|$

\begin{align*}
f'(x) &= \lim_{h \rightarrow 0} \frac{f(x + h) - f(x)}{h} \\\\
&\text{or} \\
f'(x) &= \lim_{h \rightarrow 0} \frac{f(x + h) - f(x - h)}{2h} \\\\
\end{align*}

```python
def differenciable(x):
    return x**2

def non_differenciable(x):
    return np.abs(x)

def derivative(f, x):
    h = 1e-5
    return (f(x + h) - f(x - h)) / (2.0 * h)
```

```python
x = 1
h = 1e-10
f_x = differenciable(x)
f_xh = differenciable(x + h)

df_x = (f_xh - f_x) / h
print(df_x)
```

```python
def plot_differenciation(func):
    x = np.arange(-10, 10.1, 0.1)
    h = 1e-5
    f_x = func(x)
    f_xh = func(x + h)

    df_x = (f_xh - f_x) / h

    fig, ax = plt.subplots(1, 2, figsize=(5, 2))
    ax[0].plot(x, f_x, label='f(x)')
    ax[0].plot(x, f_xh, label='f(x+h)')
    # Plot the derivative
    ax[1].plot(x, df_x, label='df(x)/dx')
    fig.tight_layout()
    fig.show();
```

```python
plot_differenciation(differenciable)
```

```python
plot_differenciation(non_differenciable)
```
